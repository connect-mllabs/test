```python
import pandas  as pd
data=np.random.randint(0,100,100)
```

```python
data[-3:]=[1000,1000,1200]
```

```python
fqrt=np.quantile(data,0.25)
tqrt=np.quantile(data,0.75)
sqrt=np.quantile(data,0.50)
iqr=tqrt-fqrt

print(fqrt,tqrt,sqrt,iqr)

ltv=fqrt-1.5*(iqr)
utv=tqrt+1.5*(iqr)

ltv

utv

for i in data:
    if i<ltv or i>utv:
        print("found an outlier",i)
```
